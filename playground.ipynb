```python
import numpy as np
import scipy.sparse as sp
import implicit

with open('data/vi_dataset_events.csv', 'r') as f:
    data = f.readlines()

d = np.zeros((86016, 28369))

for line in data[1:]:
    user, item, event, timestamp = line.split(',')
    user = np.int8(user)
    item = np.int8(item)
    print(line.split(','))
    if event == 'purchase':
        weight = 100
    elif event == 'add_to_cart':
        weight = 10
    else:
        weight = 1

    d[user, item] = 1

print(d)
mat = sp.csr_matrix(d)
print(mat)
model = implicit.als.AlternatingLeastSquares(factors=16)
model.fit(mat)
print(model)
user = d[1]
print(user)
```

```python
with open('data/vi_dataset_events.csv', 'r') as f:
    data = f.readlines()
    
mat = sp.csr_matrix((86016, 28369), dtype=np.int8)

for line in data[1:]:
    user, item, event, timestamp = line.split(',')
    user = np.int8(user)
    item = np.int8(item)
    
    if event == 'purchase':
        weight = 100
    elif event == 'add_to_cart':
        weight = 10
    else:
        weight = 1
        
    mat[user, item] += weight  

model = implicit.als.AlternatingLeastSquares(factors=50)

model.fit(mat)
```
